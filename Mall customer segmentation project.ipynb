```python
Score_1_25=df['Score'][(df["Score"] >=1) & (df["Score"]<=25)]
Score_26_50=df["Score"][(df["Score"] >= 26) & (df["Score"] <= 50)]
Score_51_75=df["Score"][(df["Score"] >=51) & (df["Score"] <= 75)]
Score_76_100=df["Score"][(df["Score"] >=76) & (df["Score"] <= 100)]

score_x=["1-25","26-50","51-75","76-100"]
score_y=[len(Score_1_25.values), len(Score_26_50.values),len(Score_51_75.values),len(Score_76_100.values)]

plt.figure(figsize=(5,5))
sx=sns.barplot(x=score_x,y=score_y,data=df)
#plt.title("Spending Score")
#plt.xlabel("Score")
#plt.ylable("Customers")
sx.set(xlabel="Score",ylabel="Customers",title="Spending Score")
plt.show()
```

```python
q=df.loc[:,["Age","Score"]].values
from sklearn.cluster import KMeans
a=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++")
    kmeans.fit(q)
    a.append(kmeans.inertia_)
plt.figure(figsize=(5,5))
plt.grid()
plt.plot(range(1,11),a,color="b",lw=2,marker=10)
plt.xlabel("Kmeans")
plt.ylabel("clusters")
plt.show()
```

```python
KM=KMeans(n_clusters=4) #intitalize the class object
pr=KM.fit_predict(q) #predict the labels of clusters
pr
```

```python
KM.cluster_centers_
```

```python
plt.scatter(q[:,0],q[:,1],c=KM.labels_,cmap='plasma')
plt.scatter(KM.cluster_centers_[:,0],KM.cluster_centers_[:,1],color='r',s=100)
plt.title('clustering of data ')
plt.xlabel('Age')
plt.ylabel('Score')
plt.grid()
plt.show()
```

```python
q1=df.loc[:,["Annual Income (k$)","Score"]].values
from sklearn.cluster import KMeans
a=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init="k-means++")
    kmeans.fit(q1)
    a.append(kmeans.inertia_)
plt.figure(figsize=(5,5))
plt.grid()
plt.plot(range(1,11),a,color="b",lw=2,marker=10)
plt.xlabel("Kmeans")
plt.ylabel("clusters")
plt.show()
```
